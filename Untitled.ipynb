Step 3:KMeans 

```python
# Extracting the values from the 6th (index 5) and 10th (index 9) columns of the DataFrame into a NumPy array (X). 
# These columns are the features that will be used for clustering, such as 'Purchase Amount (GBP)' and 'Review Ratings'.
X = df.iloc[:, [5, 9]].values 
```

```python
# Displaying the extracted values from the selected columns (X) to preview the features that will be used in the clustering process.
X
```

```python
cluster = [] # Initializing an empty list to store the inertia values for each number of clusters.
for i in range(1, 11):  # Looping through cluster numbers from 1 to 10.
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=0) # Initializing KMeans with i clusters, using k-means++ initialization.
    kmeans.fit(X) # Fitting the KMeans model to the data (X).
    cluster.append(kmeans.inertia_)  # Appending the inertia value (sum of squared distances) to the cluster list for each iteration.
```

```python
plt.plot(range(1, 11), cluster)  # Plotting the number of clusters (1 to 10) against the inertia values.
plt.title('The Elbow Method') # Adding a title to the plot.
plt.xlabel('Number of clusters') # Labeling the x-axis as 'Number of clusters'.
plt.ylabel('cluster') # Labeling the y-axis as 'Cluster' (inertia values).
plt.show() # Displaying the plot.
```

```python
kmeansmodel = KMeans(n_clusters= 3, init='k-means++', random_state=0) # Initializing the KMeans model with 3 clusters, using 'k-means++' initialization to speed up convergence, and setting a random state for reproducibility
y_kmeans = kmeansmodel.fit_predict(X) # Fitting the KMeans model to the data (X) and predicting the cluster labels for each data point.
```

```python
# Plotting the scatter plot for each cluster:
# - Cluster 1 is represented by red points.
# - Cluster 2 is represented by blue points.
# - Cluster 3 is represented by green points.
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s =100, c='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s =100, c='blue', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s =100, c='green', label='Cluster 3')

# Plotting the centroids of the clusters as yellow points.
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'yellow', label='Centroids')

# Adding a title and labels to the plot.
plt.title('Clusters of Customers') # Title of the plot.
plt.xlabel('Purchase Amount (GBP)') # Label for the x-axis.
plt.ylabel('Review Rating') # Label for the y-axis.

# Displaying the legend to show labels for each cluster and the centroids.
plt.legend()

# Displaying the plot.
plt.show()
```

Step 4: Content-Based Filtering

```python
# Set display options to show all rows and columns in the dataframe
pd.set_option('display.max_rows', None)  # To show all rows
pd.set_option('display.max_columns', None)  # To show all columns

# Function to recommend items based on user-selected color and category
def recommend_items_by_color_and_category(user_color, user_category):
    # Filter the dataset for items that match the user's selected color and category
    filtered_items = df[(df['Color'].str.lower() == user_color.lower()) & (df['Category'].str.lower() == user_category.lower())]
    
    # Return the filtered DataFrame 
    return filtered_items[['Item Purchased', 'Color', 'Category']]

# Example: User asks for black items in the 'Footwear' category
recommendations = recommend_items_by_color_and_category("pink", "Footwear")
print(recommendations)
```

```python
# Function to recommend items based on user-selected item and price range
def recommend_items_by_item_and_price_range(user_item, min_price, max_price):
    # Check if the input 'user_item' is empty
    if not user_item:
        return "Please enter an item to search."
    
    # Check if the min and max prices are valid
    if min_price < 0 or max_price < 0:
        return "Price values must be positive."
    
    # Adjust price range if min_price > max_price (swap the values)
    if min_price > max_price:
        min_price, max_price = max_price, min_price
    
    # Filter the dataset for items that match the user's selected item (case-insensitive) and fall within the price range
    filtered_items = df[(df['Item Purchased'].str.contains(user_item, case=False, na=False)) &
                         (df['Purchase Amount (GBP)'] >= min_price) &
                         (df['Purchase Amount (GBP)'] <= max_price)]
    
    # Check if no items match the selected item and price range
    if filtered_items.empty:
        return f"No items found for '{user_item}' in the price range between {min_price} and {max_price}."
    
    # Sort the filtered items by price (ascending order)
    sorted_items = filtered_items.sort_values(by='Purchase Amount (GBP)', ascending=True)
    
    # Return the filtered items with 'Item Purchased' and 'Purchase Amount (GBP)'
    return sorted_items[['Item Purchased', 'Purchase Amount (GBP)']]

# Example: User asks for items containing 'T-shirt' in the price range of 20 to 10
recommendations_by_item_price_range = recommend_items_by_item_and_price_range("jeans", 30, 50)
print(recommendations_by_item_price_range)
```

```python
# Function to recommend items based on user-selected gender and category
def recommend_items_by_gender_and_category(user_gender, user_category):
    # Check if the inputs are empty
    if not user_gender or not user_category:
        return "Please enter both gender and category to search."
    
    # Filter the dataset for items that match the user's selected gender and category (case-insensitive)
    filtered_items = df[(df['Gender'].str.lower() == user_gender.lower()) &
                         (df['Category'].str.lower() == user_category.lower())]
    
    # Check if no items match the selected gender and category
    if filtered_items.empty:
        return f"No items found for gender '{user_gender}' in the '{user_category}' category."
    
    # Return the filtered items with 'Item Purchased', 'Gender', and 'Category'
    return filtered_items[['Item Purchased', 'Gender', 'Category']]

# Example: User asks for items for 'Male' gender in the 'Clothing' category
recommendations_by_gender_category = recommend_items_by_gender_and_category("Female", "Footwear")
print(recommendations_by_gender_category)
```

```python
# Function to recommend items based on user-selected item and size (exact match)
def recommend_items_by_item_and_size(user_item, user_size):
    # Strip extra spaces and ensure case-insensitive matching
    user_item = user_item.strip().lower()
    user_size = user_size.strip().lower()

    # Filter the dataset for exact matches with 'Item Purchased' and 'Size'
    filtered_items = df[(df['Item Purchased'].str.lower() == user_item) & 
                        (df['Size'].str.lower() == user_size)]
    
    # Check if any items match the selected item and size
    if filtered_items.empty:
        return f"No items found for item '{user_item}' and size '{user_size}'. Please check your input and try again."
    
    # Return the filtered items with 'Item Purchased' and 'Size' 
    return filtered_items[['Item Purchased', 'Size']]

# Example: User asks for "Shoes" in the 'L' size
recommendations_by_item_and_size = recommend_items_by_item_and_size("Shorts", "L")
print(recommendations_by_item_and_size)
```

```python
# Function to recommend items based on user-selected item and review rating (high or low)
def recommend_items_by_item_and_rating(user_item, user_rating):
    # Strip extra spaces and ensure case-insensitive matching
    user_item = user_item.strip().lower()
    user_rating = user_rating.strip().lower()

    # Filter the dataset for items that match the user's selected item
    filtered_items = df[df['Item Purchased'].str.contains(user_item, case=False, na=False)]

    # Apply filter for review rating based on user choice
    if user_rating == 'high':
        filtered_items = filtered_items[filtered_items['Review Rating'] >= 4]
    elif user_rating == 'low':
        filtered_items = filtered_items[filtered_items['Review Rating'] < 4]
    else:
        return "Please choose either 'high' or 'low' for the review rating."

    # Check if any items match the selected item and rating
    if filtered_items.empty:
        return f"No items found for item '{user_item}' with '{user_rating}' review rating. Please check your input and try again."

    # Sort the filtered items by Review Rating (high to low or low to high based on user preference)
    sorted_items = filtered_items.sort_values(by='Review Rating', ascending=False if user_rating == 'high' else True)
    
    # Return the sorted items with 'Item Purchased' and 'Review Rating'
    return sorted_items[['Item Purchased', 'Review Rating']]

# Example: User asks for items like "Shoes" with a high review rating
recommendations_by_item_and_rating = recommend_items_by_item_and_rating("Jeans", "high")
print(recommendations_by_item_and_rating)
```

```python
# Function to recommend items based on user-selected item and age range
def recommend_items_by_item_and_age_range(user_item, min_age, max_age):
    # Strip extra spaces and ensure case-insensitive matching
    user_item = user_item.strip().lower()

    # Filter the dataset for items that match the user's selected item
    filtered_items = df[df['Item Purchased'].str.contains(user_item, case=False, na=False)]

    # Ensure Age column is numeric (if not, convert it to NaN)
    filtered_items.loc[:, 'Age'] = pd.to_numeric(filtered_items['Age'], errors='coerce')

    # Apply the age filter on filtered_items (not df)
    filtered_items = filtered_items[(filtered_items['Age'] >= min_age) & (filtered_items['Age'] <= max_age)]
    
    # Check if any items match the selected item and age range
    if filtered_items.empty:
        return f"No items found for item '{user_item}' and age range '{min_age}-{max_age}'. Please check your input and try again."

    # Sort the filtered items by Age (low to high) to get the age range in the correct order
    sorted_items = filtered_items.sort_values(by='Age', ascending=True)
    
    # Return the filtered items with 'Item Purchased' and 'Age'
    return sorted_items[['Item Purchased', 'Age']]

# Example: User asks for items like "Jeans" for the age range 20 to 60
recommendations_by_item_and_age_range = recommend_items_by_item_and_age_range("Skirt", 20, 60)
print(recommendations_by_item_and_age_range)
```
